# Automatic Flake Detector

This python program was created and tested in python version 3.10.11.

Other python versions may not be compatible.

```python
import main
import os
import motor_functions as mf
import capture_functions as cf
from display_flakes import Display
import map
```

## Set parameters.

    1. Set target folder.

    2. Set positions to be scanned in the wafer tray.

    3. Set false positive probability.

```python
target_folder = "C:/Users/Graph/OneDrive/Desktop/automated microscope/trials/0820_2"

variables = {
    (2, 0): 0, (2, 1): 0, (2, 2): 0, 
    (1, 0): 0, (1, 1): 1, (1, 2): 0, 
    (0, 0): 0, (0, 1): 0, (0, 2): 0
}

probability = 0.6 
```

Create input and output folders within the given folder and initialize the motors.

```python
input_folder = os.path.join(target_folder, f"input")
output_folder = os.path.join(target_folder, f"output")

os.makedirs(input_folder, exist_ok = True)
os.makedirs(output_folder, exist_ok = True)

mf.initialize()
```

## Before detecting, you have to: 
    1. CHANGE THE "SAVE IN" DIRECTORY in Remote APP to the input folder within your target folder. (VERY IMPORTANT)

    2. Make sure the Remote APP covers the LEFT half of the screen.

    3. Push the wafer to the left-bottom corner, and push the wafer tray to the right-bottom corner.

Once done, you may start detecting flakes. 

You will be notified when the program has finished detecting flakes when a popup appears on the screen notifying you that capturing and detecting flakes has finished.

If you want the program to terminate early, move the mouse cursor to the left-bottom corner of the screen. Wait until a popup appears notifying you that capturing and detecting flakes has finished. This indicates the program has ended. After the program has ended, check the input and output folders. 

```python
main.Main(probability, target_folder, input_folder, output_folder, variables)
```

## Show all detected flakes.

Running this will produce a window that will display images of the detected flakes. Make sure it is VISIBLE ALL THE TIME because it is needed to slightly adjust the position of the stage.

Press:

    - "l" to go to the next image 

    - "j" to go back one image 

    - "k" to move the motor to the flake in the picture 

    - "a", "d", "w", "s" to move a small step left, right, up or down

    - "0" to delete the current flake

    - "Esc" to exit the window

Here you can switch to 100x to take a close look at the flakes.
