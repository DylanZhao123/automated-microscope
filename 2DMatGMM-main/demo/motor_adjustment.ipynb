# Stage Position Adjustment

If the stage failed to move back to the origin after the main program terminates, you can use this program to move it manually back to the origin.

```python
import cv2
import motor_functions as mf

mf.initialize()

# Create an OpenCV window for adjustments
cv2.namedWindow("adjustment", cv2.WINDOW_NORMAL)
cv2.setWindowTitle("adjustment", "stage adjustment")

X, Y = 0, 0

while True:
    # Wait for a key press for 10 ms
    key = cv2.waitKey(10)

    # Break the loop if the Escape key (27) is pressed
    if key == 27:
        break

    # Check for 'a' key press (move left)
    if key == ord("a"):
        mf.move_to(X, Y - 0.1)
        Y = Y - 0.1

    # Check for 'w' key press (move up)
    if key == ord("w"):
        mf.move_to(X + 0.1, Y)
        X = X + 0.1

    # Check for 'd' key press (move right)
    if key == ord("d"):
        mf.move_to(X, Y + 0.1)
        Y = Y + 0.1

    # Check for 's' key press (move down)
    if key == ord("s"):
        mf.move_to(X - 0.1, Y)
        X = X - 0.1

# Close OpenCV windows and clean up motor functions when finished
cv2.destroyAllWindows()
#mf.cleanup()
```
